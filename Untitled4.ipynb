```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load data
df = pd.read_csv('/content/S08A1_livestock.csv')

print("Dataset Shape:", df.shape)
print("\nColumn Names:")
print(df.columns.tolist())
print("\nData Types:")
print(df.dtypes)
print("\nMissing Values:")
print(df.isnull().sum())
print("\nFirst few rows:")
df.head()
```

```python
original_shape = df.shape
print(f"Original dataset shape: {original_shape}")

# Separate numeric and non-numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
non_numeric_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()

print(f"\nNumeric columns ({len(numeric_cols)}): {numeric_cols}")
print(f"Non-numeric columns ({len(non_numeric_cols)}): {non_numeric_cols}")

df_numeric = df[numeric_cols].copy()

print(f"\nMissing values before imputation:")
print(df_numeric.isnull().sum())

# Impute missing values with mean
df_numeric = df_numeric.fillna(df_numeric.mean())

print(f"\nMissing values after imputation:")
print(df_numeric.isnull().sum())

# Remove columns with zero variance
df_numeric = df_numeric.loc[:, df_numeric.std() != 0]

print(f"\nFinal shape: {df_numeric.shape}")

data = df_numeric.values
print(f"Data array shape: {data.shape}")
```

```python
# Calculate mean and std
mean = np.mean(data, axis=0)
std = np.std(data, axis=0, ddof=1)

print(f"Original mean range: [{mean.min():.2f}, {mean.max():.2f}]")
print(f"Original std range: [{std.min():.2f}, {std.max():.2f}]")

# Standardize: z = (x - μ) / σ
standardized_data = (data - mean) / std

print(f"\nAfter standardization:")
print(f"Mean: {np.mean(standardized_data, axis=0)}")
print(f"Std: {np.std(standardized_data, axis=0, ddof=1)}")

# Display first 5 rows
standardized_data[:5]
```

```python
n_samples = standardized_data.shape[0]
n_features = standardized_data.shape[1]

# Calculate covariance matrix
cov_matrix = (1 / (n_samples - 1)) * np.dot(standardized_data.T, standardized_data)

print(f"Covariance matrix shape: {cov_matrix.shape}")
print(f"Is symmetric: {np.allclose(cov_matrix, cov_matrix.T)}")

# Visualize
plt.figure(figsize=(10, 8))
sns.heatmap(cov_matrix,
            xticklabels=df_numeric.columns,
            yticklabels=df_numeric.columns,
            annot=True if n_features <= 10 else False,
            fmt='.2f',
            cmap='coolwarm',
            center=0,
            square=True)
plt.title('Covariance Matrix')
plt.tight_layout()
plt.show()

cov_matrix
```

```python
# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print(f"Number of eigenvalues: {len(eigenvalues)}")
print(f"\nEigenvalues:\n{eigenvalues}")
print(f"\nSum of eigenvalues: {np.sum(eigenvalues):.4f}")
print(f"Trace of covariance matrix: {np.trace(cov_matrix):.4f}")

eigenvalues, eigenvectors
```

```python
# Sort in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

print("Sorted eigenvalues:")
for i, val in enumerate(sorted_eigenvalues):
    print(f"PC{i+1}: {val:.4f}")

# Calculate explained variance
explained_variance = sorted_eigenvalues / np.sum(sorted_eigenvalues)
cumulative_variance = np.cumsum(explained_variance)

print(f"\n{'Component':<12} {'Eigenvalue':<12} {'Variance %':<12} {'Cumulative %':<12}")
for i in range(len(explained_variance)):
    print(f"PC{i+1:<10} {sorted_eigenvalues[i]:<12.4f} {explained_variance[i]*100:<12.2f} {cumulative_variance[i]*100:<12.2f}")

# Plot
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Scree plot
axes[0].bar(range(1, len(sorted_eigenvalues)+1), sorted_eigenvalues, alpha=0.7)
axes[0].plot(range(1, len(sorted_eigenvalues)+1), sorted_eigenvalues, 'ro-')
axes[0].set_xlabel('Principal Component')
axes[0].set_ylabel('Eigenvalue')
axes[0].set_title('Scree Plot')
axes[0].grid(True, alpha=0.3)

# Explained variance
axes[1].bar(range(1, len(explained_variance)+1), explained_variance * 100, alpha=0.7)
axes[1].set_xlabel('Principal Component')
axes[1].set_ylabel('Explained Variance (%)')
axes[1].set_title('Explained Variance per Component')
axes[1].grid(True, alpha=0.3)

# Cumulative variance
axes[2].plot(range(1, len(cumulative_variance)+1), cumulative_variance * 100, 'go-')
axes[2].axhline(y=95, color='r', linestyle='--', label='95% threshold')
axes[2].axhline(y=90, color='orange', linestyle='--', label='90% threshold')
axes[2].set_xlabel('Number of Components')
axes[2].set_ylabel('Cumulative Explained Variance (%)')
axes[2].set_title('Cumulative Explained Variance')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

sorted_eigenvectors
```

```python
# Select number of components to retain 95% variance
threshold = 0.95
cumulative = np.cumsum(explained_variance)
num_components = np.argmax(cumulative >= threshold) + 1

print(f"Components needed for {threshold*100}% variance: {num_components}")
print(f"Actual variance captured: {cumulative[num_components-1]*100:.2f}%")
print(f"Dimensionality reduction: {n_features} → {num_components}")

# Get top eigenvectors
top_eigenvectors = sorted_eigenvectors[:, :num_components]

# Project data
reduced_data = np.dot(standardized_data, top_eigenvectors)

print(f"\nOriginal shape: {standardized_data.shape}")
print(f"Reduced shape: {reduced_data.shape}")

reduced_data[:5]
```

```python
print(f'Reduced Data Shape: {reduced_data.shape}')
print(f'Features: {standardized_data.shape[1]} → {reduced_data.shape[1]}')
print(f'Variance retained: {cumulative[num_components-1]*100:.2f}%')

# Create dataframe
reduced_df = pd.DataFrame(
    reduced_data,
    columns=[f'PC{i+1}' for i in range(num_components)]
)

print("\nStatistics:")
print(reduced_df.describe())

reduced_df.head()
```

```python
fig = plt.figure(figsize=(16, 10))

# Original data - first 2 features
ax1 = plt.subplot(2, 3, 1)
ax1.scatter(standardized_data[:, 0], standardized_data[:, 1],
           alpha=0.6, c=reduced_data[:, 0], cmap='viridis', s=50)
ax1.set_xlabel(f'{df_numeric.columns[0]}')
ax1.set_ylabel(f'{df_numeric.columns[1]}')
ax1.set_title('Original Data (First 2 Features)')
ax1.grid(True, alpha=0.3)

# PCA - PC1 vs PC2
ax2 = plt.subplot(2, 3, 2)
ax2.scatter(reduced_data[:, 0], reduced_data[:, 1],
           alpha=0.6, c=reduced_data[:, 0], cmap='plasma', s=50)
ax2.set_xlabel(f'PC1 ({explained_variance[0]*100:.1f}%)')
ax2.set_ylabel(f'PC2 ({explained_variance[1]*100:.1f}%)')
ax2.set_title('After PCA (PC1 vs PC2)')
ax2.grid(True, alpha=0.3)

# 3D if available
if num_components >= 3:
    from mpl_toolkits.mplot3d import Axes3D
    ax3 = plt.subplot(2, 3, 3, projection='3d')
    ax3.scatter(reduced_data[:, 0], reduced_data[:, 1], reduced_data[:, 2],
               c=reduced_data[:, 0], cmap='coolwarm', alpha=0.6, s=50)
    ax3.set_xlabel(f'PC1 ({explained_variance[0]*100:.1f}%)')
    ax3.set_ylabel(f'PC2 ({explained_variance[1]*100:.1f}%)')
    ax3.set_zlabel(f'PC3 ({explained_variance[2]*100:.1f}%)')
    ax3.set_title('3D PCA Projection')

# Feature importance in PC1
ax4 = plt.subplot(2, 3, 4)
feature_importance = np.abs(sorted_eigenvectors[:, 0])
sorted_idx = np.argsort(feature_importance)[::-1]
ax4.barh(range(len(feature_importance)), feature_importance[sorted_idx])
ax4.set_yticks(range(len(df_numeric.columns)))
ax4.set_yticklabels([df_numeric.columns[i] for i in sorted_idx], fontsize=8)
ax4.set_xlabel('Absolute Loading')
ax4.set_title('Feature Importance in PC1')
ax4.grid(True, alpha=0.3, axis='x')

# Biplot
ax5 = plt.subplot(2, 3, 5)
ax5.scatter(reduced_data[:, 0], reduced_data[:, 1], alpha=0.5, s=30, c='gray')

scale = 3
for i in range(min(n_features, 10)):
    ax5.arrow(0, 0,
             sorted_eigenvectors[i, 0] * scale,
             sorted_eigenvectors[i, 1] * scale,
             head_width=0.1, head_length=0.1, fc='red', ec='red', alpha=0.7)
    ax5.text(sorted_eigenvectors[i, 0] * scale * 1.15,
            sorted_eigenvectors[i, 1] * scale * 1.15,
            df_numeric.columns[i], fontsize=8, ha='center')

ax5.set_xlabel(f'PC1 ({explained_variance[0]*100:.1f}%)')
ax5.set_ylabel(f'PC2 ({explained_variance[1]*100:.1f}%)')
ax5.set_title('Biplot')
ax5.grid(True, alpha=0.3)
ax5.axhline(0, color='black', linewidth=0.5)
ax5.axvline(0, color='black', linewidth=0.5)

# PC distributions
ax6 = plt.subplot(2, 3, 6)
for i in range(min(3, num_components)):
    ax6.hist(reduced_data[:, i], bins=30, alpha=0.5, label=f'PC{i+1}')
ax6.set_xlabel('Score Value')
ax6.set_ylabel('Frequency')
ax6.set_title('Distribution of PC Scores')
ax6.legend()
ax6.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import time
from sklearn.decomposition import PCA

# Benchmark custom implementation
start = time.time()
mean_b = np.mean(data, axis=0)
std_b = np.std(data, axis=0, ddof=1)
standardized_b = (data - mean_b) / std_b
cov_b = (1 / (data.shape[0] - 1)) * np.dot(standardized_b.T, standardized_b)
eigenvalues_b, eigenvectors_b = np.linalg.eig(cov_b)
sorted_idx_b = np.argsort(eigenvalues_b)[::-1]
top_eigenvectors_b = eigenvectors_b[:, sorted_idx_b[:num_components]]
reduced_b = np.dot(standardized_b, top_eigenvectors_b)
your_time = time.time() - start

# Benchmark sklearn
start = time.time()
pca_sklearn = PCA(n_components=num_components)
sklearn_reduced = pca_sklearn.fit_transform(standardized_data)
sklearn_time = time.time() - start

print(f"Custom implementation: {your_time:.6f} seconds")
print(f"Sklearn implementation: {sklearn_time:.6f} seconds")
print(f"Speed ratio: {your_time/sklearn_time:.2f}x")

# Verify correctness
results_match = np.allclose(np.abs(reduced_data), np.abs(sklearn_reduced), rtol=1e-3)
print(f"\nResults match sklearn: {results_match}")

print(f"\nCustom explained variance: {explained_variance[:num_components]}")
print(f"Sklearn explained variance: {pca_sklearn.explained_variance_ratio_}")

# Compare visually
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].scatter(reduced_data[:, 0], reduced_data[:, 1], alpha=0.6, c='blue')
axes[0].set_xlabel('PC1')
axes[0].set_ylabel('PC2')
axes[0].set_title('Custom PCA')
axes[0].grid(True, alpha=0.3)

axes[1].scatter(sklearn_reduced[:, 0], sklearn_reduced[:, 1], alpha=0.6, c='red')
axes[1].set_xlabel('PC1')
axes[1].set_ylabel('PC2')
axes[1].set_title('Sklearn PCA')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
